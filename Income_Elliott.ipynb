```python
#Load states shapefile (for plotting) (https://www.census.gov/geographies/mapping-files/time-series/geo/carto-boundary-file.html under States)

statesShp = gp.read_file(dataFolder + "/cb_2018_us_state_500k.zip").to_crs(3857)
statesShp = statesShp[['GEOID', 'STUSPS', 'NAME', 'geometry']]
TXShp = statesShp[statesShp['NAME'] == 'Texas']

def plotUS():
    ax = statesShp.plot(figsize=(20, 10))
    ax.axis('equal')
    plt.xlim(-1.5e7, -0.7e7)
    plt.ylim(2.5e6, 7e6)
    return ax

def plotTX():
    ax = TXShp.plot(figsize=(20, 14), color = 'lightgrey')
    ax.axis('equal')
    plt.xlim(-1.2e7, -1.0e7)
    plt.ylim(2.8e6, 4.6e6)
    return ax
```
